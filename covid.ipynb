```python
ages = []
for a in new_df['age']:
    try:
        a = int(a)
        
    except ValueError:
        pass
    else:
        ages.append(a)
        
age = [i for i in range(1, 81)]
m_age = [ages.count(v) for v in age]
plt.style.use('seaborn')
fig, ax = plt.subplots(figsize=(20, 10))
ax.plot(age, m_age, c='red')
ax.set_xlabel('Age', fontsize=20)
ax.set_ylabel('Cases Acc. to Age', fontsize=20)
ax.set_title('Covid With Ages', fontsize=25)
ax.tick_params(axis='both', labelsize=15)
plt.show()
```

```python
from plotly.graph_objs import Scattergeo, Layout
from plotly import offline


lati = new_df['latitude']
logi = new_df['longitude']
latis = []
logis = []
cities = [c for c in new_df['city']]
countries = [c for c in new_df['country']]
z = zip(cities, countries)  
title = list(z)


for l in lati:
    try:
        l = float(l)
    except ValueError:
        pass
    else:
        latis.append(l)


for l in logi:
    try:
        l = float(l)
    except ValueError:
        pass
    else:
        logis.append(l)


data = {
    'type': 'scattergeo',
    'lon': logis,
    'lat': latis,
    'text': title,
    'marker': {
        'color': 'rgb(50, 100, 10)',
        'colorscale': 'Viridis',
        'reversescale': True,
        'colorbar': {'title': 'Magnitude'},
    }
}
layout = Layout(title='Covid19 World Map')
fig = {'data': data, 'layout': layout}
offline.plot(fig, filename='covid_map.html')
```
